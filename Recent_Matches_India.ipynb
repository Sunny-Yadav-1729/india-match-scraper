```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
import time


opts = Options()
opts.add_argument("--headless")
opts.add_argument("--disable-gpu")
opts.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) ...")

driver = webdriver.Chrome(options=opts)
driver.get("https://www.cricbuzz.com/cricket-team/india/2/results")
time.sleep(5)  # wait for JS to fully load

html = driver.page_source
driver.quit()

soup = BeautifulSoup(html, "html.parser")
```

```python
# print(soup.prettify())
frame = soup.find_all("div",class_ = "cb-col-67 cb-col cb-left cb-hm-rght")[0]
# print(frame.find_all("div",class_="cb-col-100 cb-col cb-brdr-thin-btm")[0])
# print(frame.find_all("div",class_="cb-col-100 cb-col cb-brdr-thin-btm")[0].text)
frame = frame.find_all("div",class_="cb-col-100 cb-col cb-brdr-thin-btm")
frame[0].find_all("a")[1].text
# frame[0]
```

```python
import pandas as pd
Dates = []
Team = []
Win_Team = []
Stadium = []
League = []
df = pd.DataFrame()
# frame = frame[2:]
for f in frame :
      # print(f)
# date
# print(f.find_all("div",class_="cb-col-100 cb-col cb-brdr-thin-btm")[0].find_all("span")[0].text)
      Dates.append(f.find_all("span")[0].text)
# # team
     
      Team.append(f.find_all("span")[1].text)
# # win
# # print(f.find_all("a",class_="cb-text-complete")[0].text)
      Win_Team.append(f.find_all("a")[1].text)
# # leage
# # print(f.find_all("div",class_="cb-col-100 cb-col cb-brdr-thin-btm")[0].find_all("div")[1].text)
      League.append(f.find_all("div")[1].text)
# # stadium
# # print(f.find_all("div",class_="cb-col-100 cb-col cb-brdr-thin-btm")[0].find_all("div")[2].text)
      Stadium.append(f.find_all("div")[2].text)
      Temp = pd.DataFrame({"Date" : Dates,"Teams" : Team ,"Win_Team" : Win_Team,"League":League,"Stadiun":Stadium})
      df = df.concat(Temp,ignore_index = True)
```
